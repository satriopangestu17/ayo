```python
#CONTENT BASED FILTERING

import pandas as pd

#dataframe
df_books=pd.read_csv('books.csv')
df_ratings = pd.read_csv('ratings.csv')

## dengan menggunakan fungsi tidak akan muncul 0 diawalnya
def mergecol(i):
    return str(i['authors'])+' '+str(i['original_title'])+' '+str(i['title'])+' '+str(i['language_code'])
# type(mergecol(df_books))
# df_books['feature'] = df_books.apply(mergecol,axis='columns')
df_books['feature'] = (df_books['authors'])+' '+(df_books['original_title'])+' '+(df_books['title'])+' '+(df_books['language_code'])
# str(df_books['authors'])  ## kalau pakai str jadi ada ngikut 0 nya \n \n2 dll..

df_books


```

```python

df = pd.DataFrame([[1,2],[2,5]], columns=['a', 'b'])
df

def coba(x):
#     return str(x[['a']])*4   ##[[]] ; menandakan bahwa sudah berbentuk dataframe
    return str(x['a']*4)
df['baru'] = df.apply(coba, axis=1)  ## kalo ga axis=1; akan error karena hanya bisa terjadi di column df['a']
df
```

```python
## BIKIN FEATURENYA DULU UNTUK TAU YNG SIMMILAR ITU YANG MANA AJA##
## gabungin semuanya di satu column
def mergecol(i):
    return str(i['authors'])+' '+str(i['original_title'])+' '+str(i['title'])+' '+str(i['language_code'])
df_books['feature'] = df_books.apply(mergecol,axis='columns')
df_books
```

```python
# count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
model = CountVectorizer(
    tokenizer=lambda i:i.split(' '),
    analyzer='word')
matrixFeature = model.fit_transform(df_books['feature'])  ##buat masuk ke cosinus simmilarity

feature= model.get_feature_names()
# jml_fitur = len(feature)
feature
```

```python
# cosinus similarity = nilai kemiripan untuk setiap index terhadap index lain
from sklearn.metrics.pairwise import cosine_similarity
skor = cosine_similarity(matrixFeature)
# skor.tolist()[3]
skor.tolist()[3]
# skor
```
